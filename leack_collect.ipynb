```python

## Create an ArgumentParser object
#parser = argparse.ArgumentParser(description='Import and analyze leakage files')
#parser.add_argument('-s', '--subject', type=int, required=True, help='subject index (zero-based integer)')
#parser.add_argument('--suffix', type=str, default='', help='suffix of the leakage files')
#parser.add_argument('--plot', action='store_true', help='plot leakage statistics')
#args = parser.parse_args()

plot = 0
suffix = '_remove_leak'
#suffix = '_small_window'

# Paths
path_results = os.path.expandvars('$DEMARCHI_DATA_PATH') + '/results'
```

```python
participants = [f for f in os.listdir(path_results) if f[:1] == '1']
sind = 0
participant = participants[sind]
```

```python
from base import printLeakInfo
from datetime import datetime
import pandas as pd
subj2cond2avcpt = {}
df = []
#mode = 'mean'
mode = 'mean_global'
for participant in participants:
    print(participant)

    results_folder = op.join(path_results, participant, 'reorder_random')
    leakage_file = op.join(results_folder, f'leakage{suffix}.npz')

    row = {}
    dtstr = datetime.fromtimestamp(os.stat(leakage_file).st_mtime)
    row['dt'] = dtstr
    row['leakage_file'] = leakage_file
    row['participant'] = participant
    df += [row]

    print(f"Loading leakage data from: {leakage_file}")

    # Load the leakage data
    if not op.exists(leakage_file):
        print(f"Error: Leakage file {leakage_file} does not exist!")
        exit(1)

    cond2m = np.load(leakage_file, allow_pickle=True)
    #leakage_data = np.load(leakage_file)
    cond2m = cond2m['arr_0'][()]

    cond2avcpt = printLeakInfo(cond2m,  print_per_fold=0, mode=mode)
    subj2cond2avcpt[participant] = cond2avcpt
    #test_inds_clean = test_inds[ np.max(m, axis=0) <= 2 ]
df = pd.DataFrame(df)
```

```python
cond2avcpt_arr = {}
for cond,avcpt in cond2avcpt.items():
    cond2avcpt_arr[cond] = []
    
    for participant,cond2avcpt in subj2cond2avcpt.items():
        cond2avcpt_arr[cond] += [ cond2avcpt[cond] ]

print('Average across subjects of proportions (percentanges) of leakages in \n'
'initial test fold sizes (mean across folds)')
for cond,arr in cond2avcpt_arr.items():
    if cond.find('_clean') != -1:
        continue
    me = np.mean(arr)
    std = np.std(arr)
    print(f'{cond:20}: {me:5.3f}% +- {std:.3f}%')
print('Average across subjects of proportions (percentanges) of test fold sizes in \n'
'initial test fold sizes after explicit removal of timebins with leackage from test folds (mean across folds)')
for cond,arr in cond2avcpt_arr.items():
    if cond.find('_clean') == -1:
        continue
    me = np.mean(arr)
    std = np.std(arr)
    print(f'{cond:20}: {me:5.3f}% +- {std:.3f}%')
```
